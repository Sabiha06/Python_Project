```python
#1 load datset
import pandas as pd
import numpy as np
df = pd.read_csv("Office dataset.csv")
# df.set_index('ID', inplace=True, drop=True)
df.head()
```

```python
#2 identifying missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
#3 cleaning numeric columns
def clean(df):
    df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
    df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
    median_age = df['Age'].median()
    median_salary = df['Salary'].median()
    df['Age'].fillna(median_age, inplace=True)
    df['Salary'].fillna(median_salary, inplace=True)
    return df
cleaned_df = clean(df)
cleaned_df
```

```python
#4 Correcting Date Formats

df['Join Date'] = df['Join Date'].apply(lambda x: pd.to_datetime(x, errors='coerce'))
frequent_date = df['Join Date'].mode()[0]

# Fill NaN values with frequent date
df['Join Date'].fillna(frequent_date, inplace=True)

# Count unique dates 
count_unique_dates = df['Join Date'].nunique()

print(count_unique_dates)
```

```python
#5 Splitting Complex Columns
df[['CI_Part1', 'CI_Part2', 'CI_Part3']] = df['Complex Info'].str.split('-', expand=True)

# Display the resulting DataFrame
desired_columns = ['ID','Age', 'Salary','Employment Type','Join Date','Complex Info', 'CI_Part1', 'CI_Part2', 'CI_Part3','Comments', 'Address']  # Specify the columns in the desired order
df = df[desired_columns]
df.head(10)
```

```python
#6 checking the unique values and handling the misplaced values
df['Comments'].unique()

# Swap values between 'Employment Type' and 'Comments' columns based on the condition
condition = df['Comments'].isin(['Full-time', 'Part-time', 'Contract', 'Freelance', 'nan'])
df.loc[condition, ['Employment Type', 'Comments']] = df.loc[condition, ['Comments', 'Employment Type']].values
df.head(10)
```

```python
#7 cleaning and extracting address information with function

def split_address_column(df):
    # Create a copy of the DataFrame to avoid modifying the original DataFrame
    df_copy = df.copy()
    
    # Split the 'Address' column into separate columns for 'Street', 'City', and 'Country'
    df_copy[['Street', 'City', 'Country']] = df_copy['Address'].str.split(', ', expand=True)
    
    return df_copy

# Example usage:
# Assuming df is your original DataFrame
df_copy = df.copy()  # Copy the original DataFrame
re_df = split_address_column(df_copy)  # Apply the function to the copied DataFrame

# Display the copied DataFrame with the split address columns
re_df
```

```python
#8 Removing outliers
percentile_1st = np.percentile(df['Salary'], 1)
percentile_99th = np.percentile(df['Salary'], 99)

# identify and remove rows that have salaries outside the 1st and 99th percentiles
filtered_data = df[(df['Salary'] >= percentile_1st) & (df['Salary'] <= percentile_99th)]

# filtered DataFrame
filtered_data
```

```python
#9 Creating a Cleaned Summary
# Number of rows and columns
num_rows, num_cols = df.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_cols}")
# Number of unique values in each column
unique_values = df.nunique()
print("\nNumber of unique values in each column:",unique_values)
# Basic statistics for numerical columns 
age_salary_stats = df[['Age', 'Salary']].describe()
print("\nBasic statistics for 'Age' and 'Salary' columns:")
print(age_salary_stats)
```
